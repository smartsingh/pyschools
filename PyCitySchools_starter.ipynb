# PyCity Schools Analysis

* Overall, District schools perform better at reading than math. Perhaps if there is a targeted effort to help math students, District school pass rates will come more in line with Charter school pass rates.

* The best performing school spends less money than the worst performing school, indicating that budget may not be 
important to helping students succeed.

* The difference between small and medium size schools in terms of pass rate is negligible, whereas the difference in pass rate between medium size and large schools is large. This may be due to classroom size, but more research would be needed.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
sdc = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

sdc.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = sdc["school_name"].nunique()

total_students = int(sdc["student_name"].count())

budgets = sdc["budget"].unique()

total_budget = int(sum(budgets))

avg_math = sdc["math_score"].mean()

avg_read = sdc["reading_score"].mean()

over_pass = (avg_math + avg_read)/2

math_pass = len(sdc[sdc["math_score"] >= 70])/total_students

math_pass = math_pass * 100

read_pass = len(sdc[sdc["reading_score"] >= 70])/total_students

read_pass = read_pass * 100

district_df = pd.DataFrame({"Total Schools": total_schools, "Total Students": total_students, \
                            "Total Budget": total_budget, "Average Math Score": avg_math, \
                            "Average Reading Score": avg_read, "% Passing Math": math_pass, \
                            "% Passing Reading": read_pass, "% Overall Passing Rate": over_pass}, index = [0])

district_df["Total Students"] = district_df["Total Students"].map("{:,}".format)

district_df["Total Budget"] = district_df["Total Budget"].map("${:,.2f}".format)

district_df["Average Math Score"] = district_df["Average Math Score"].map("{:,.2f}".format)

district_df["Average Reading Score"] = district_df["Average Reading Score"].map("{:,.2f}".format)

district_df["% Passing Math"] = district_df["% Passing Math"].map("{:,.2f}%".format)

district_df["% Passing Reading"] = district_df["% Passing Reading"].map("{:,.2f}%".format)

district_df["% Overall Passing Rate"] = district_df["% Overall Passing Rate"].map("{:,.2f}%".format)

district_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
names = sdc["school_name"].unique()

schools = []

types = []

s_students = []

s_budgets = []

s_avg_math = []

s_avg_read = []

s_pass_math_count = []

s_pass_read_count = []

for name in names:
    schools.append(name)
    types.append(sdc.loc[sdc["school_name"] == name, "type"].iloc[0])
    s_budgets.append(sdc.loc[sdc["school_name"] == name, "budget"].iloc[0])
    s_students.append(len(sdc.loc[sdc["school_name"]== name]))
    s_avg_math.append(sdc.loc[sdc["school_name"] == name, "math_score"].mean())
    s_avg_read.append(sdc.loc[sdc["school_name"] == name, "reading_score"].mean())
    s_pass_math_count.append(len(sdc.loc[(sdc["school_name"] == name) & (sdc["math_score"] >=70)]))
    s_pass_read_count.append(len(sdc.loc[(sdc["school_name"] == name) & (sdc["reading_score"] >=70)]))
    
sdf_raw = pd.DataFrame({"School Type":types, "Total Students":s_students, "Total School Budget":s_budgets, \
                    "Average Math Score": s_avg_math, "Average Reading Score": s_avg_read, \
                    "Math_passes":s_pass_math_count, "Read_passes":s_pass_read_count}, index=schools)

sdf_raw["Per Student Budget"] = sdf_raw["Total School Budget"]/sdf_raw["Total Students"]

sdf_raw["% Passing Math"] = sdf_raw["Math_passes"]/sdf_raw["Total Students"] * 100

sdf_raw["% Passing Reading"] = sdf_raw["Read_passes"]/sdf_raw["Total Students"] * 100

sdf_raw["Overall Passing Rate"] = (sdf_raw["% Passing Math"]+sdf_raw["% Passing Reading"]) / 2

del sdf_raw["Math_passes"]

del sdf_raw["Read_passes"]

sdf_raw = sdf_raw[["School Type", "Total Students", "Total School Budget", "Per Student Budget", "Average Math Score", 
                  "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]

#sdf_raw["Total School Budget"] = sdf_raw["Total School Budget"].map("${:,.2f}".format)

#sdf_raw["Per Student Budget"] = sdf_raw["Per Student Budget"].map("${:,.2f}".format)

#sdf_raw["Average Math Score"] = sdf_raw["Average Math Score"].map("{:.2f}".format)

#sdf_raw["Average Reading Score"] = sdf_raw["Average Reading Score"].map("{:.2f}".format)

#sdf_raw["% Passing Math"] = sdf_raw["% Passing Math"].map("{:.2f}%".format)

#sdf_raw["% Passing Reading"] = sdf_raw["% Passing Reading"].map("{:.2f}%".format)

#sdf_raw["Overall Passing Rate"] = sdf_raw["Overall Passing Rate"].map("{:.2f}%".format)

sdf_raw
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth = []

tenth = []

eleventh = []

twelfth = []

for name in names:
    ninth.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "9th"), "math_score"].mean())
    tenth.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "10th"), "math_score"].mean())
    eleventh.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "11th"), "math_score"].mean())
    twelfth.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "12th"), "math_score"].mean())
    
math_grades = pd.DataFrame({"9th":ninth,"10th":tenth,"11th":eleventh,"12th":twelfth}, index=schools)

math_grades["9th"]=math_grades["9th"].map("{:.2f}".format)
math_grades["10th"]=math_grades["10th"].map("{:.2f}".format)
math_grades["11th"]=math_grades["11th"].map("{:.2f}".format)
math_grades["12th"]=math_grades["12th"].map("{:.2f}".format)

math_grades.sort_index()
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
ninth_r = []

tenth_r = []

eleventh_r = []

twelfth_r = []

for name in names:
    ninth_r.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "9th"), "reading_score"].mean())
    tenth_r.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "10th"), "reading_score"].mean())
    eleventh_r.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "11th"), "reading_score"].mean())
    twelfth_r.append(sdc.loc[(sdc["school_name"] == name) & (sdc["grade"] == "12th"), "reading_score"].mean())
    
read_grades = pd.DataFrame({"9th":ninth_r,"10th":tenth_r,"11th":eleventh_r,"12th":twelfth_r}, index=schools)

read_grades["9th"]=read_grades["9th"].map("{:.2f}".format)
read_grades["10th"]=read_grades["10th"].map("{:.2f}".format)
read_grades["11th"]=read_grades["11th"].map("{:.2f}".format)
read_grades["12th"]=read_grades["12th"].map("{:.2f}".format)

read_grades.sort_index()
```
